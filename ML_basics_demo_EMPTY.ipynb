### 1.4 Decision tree algorithm --- CART (Classification And Regression Tree)

It is a non-parametric supervised learning method used for classification and regression. It predicts the value of a target variable by learning decision rules inferred from the data features. It is easy to understand and interpret.

A nonparametric model does not make strong assumptions about the underlying population distribution or the relationship between variables. It is good for large datasets without prior knowledge.

Two keys:
- Top-down tree structure: root node > internal nodes > leaf nodes (decisions)
- Greedy splitting approach: at each split, the most optimal split threshold is chosen based on the maximum amount of overall error (loss/cost/error function) reduced after the split.

image.png

Popular loss functions to minimize:
- Classification: Gini Index (the probability of misclassification); Entropy (a measure of uncertainty/impurity of grouped instances)
- Regression: Mean Squared Error (MSE, a measure of the discrepancy between predicted and real values)

### 2.4 Goodness-of-fit

R-squared (0-100%): 

- A measure of goodness-of-fit. It is the square of the correlation r between the predicted and actual values.

- 100% r-squared represents a model explains 100% of the variation in the data.

"It is recommended to use R-Squared or rather adjusted R-Squared for evaluating the model performance of the regression models. This is primarily because R-Squared captures the fraction of variance of actual values captured by the regression model and tends to give a better picture of the quality of the regression model."

"... the disadvantage of using MSE than R-squared is that it will be difficult to gauge the performance of the model using MSE as the value of MSE can vary from 0 to any larger number. However, in the case of R-squared, the value is bounded between 0 and 1."
